```python
import pandas as pd
import numpy as np
import os
from tensorflow import keras
from tensorflow.keras import layers
from matplotlib import pyplot as plt



```

```python
path_model = "/home/lunet/cosoc/Desktop/orion_anomaly_pipeline/models/PU_2_full/PU/model_tr_2_it_0/"
```

```python
model = keras.models.load_model(path_model)
```

```python
def create_sequences(values, time_steps=400):
    # Generates a widows of length time_steps of a sequence

    output = [] 
    i = 0
    while i <= len(values) - time_steps:
        output.append(values[i: (i + time_steps)])
        i += 1
    return np.stack(output)


```

```python
path_to_data = "/home/lunet/cosoc/Desktop/orion_anomaly_pipeline/datasets/PU/test/N09_M07_F10_K001_1.csv"
data = pd.read_csv(path_to_data, index_col="timestamp")
proc_data = create_sequences(data, time_steps=400)
```

```python
x_train_pred = model.predict(proc_data)

x_train_pred[0]
```
